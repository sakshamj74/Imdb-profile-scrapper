```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
base_url=["https://www.imdb.com/list/ls068010962/","https://www.imdb.com/list/ls068010962/?page=2"]
name=[]      #list to store names of each star
profile=[]   #list to store links to profile of each star
for url in base_url:
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    i=soup.findAll('div',class_='lister-item mode-detail')
    for f in i:
        name.append(f.find('h3',class_="lister-item-header").find('a').text)
    j=soup.findAll('h3',class_="lister-item-header")
    for link in j:
        profile.append("https://www.imdb.com"+str(link.find('a')['href']))
```

```python
Other_Works=[None]*200
nick=[None]*200
star=[None]*200
trivia=[None]*200
trademark=[None]*200
quote=[None]*200
height=[None]*200
a_names=[None]*200
image=[None]*200
dob=[None]*200
bio=[None]*200
l=0
for i in profile:
    url=i
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    j=soup.findAll('div',class_='see-more inline canwrap')
    for k in j:
        if k.find('h4',class_='inline').text=='Other Works:':
            Other_Works[l]=k.text.replace("Other Works:","")
        
        if k.find('h4',class_='inline').text=='Star Sign:':  
            star[l]=k.text.replace("Star Sign:","")
        
        if k.find('h4',class_='inline').text=='Trivia:':  
            trivia[l]=k.text.replace("Trivia:","")
            
        if k.find('h4',class_='inline').text=='Trademark:':  
            trademark[l]=k.text.replace("Trademark:","")
        
        if k.find('h4',class_='inline').text=='Personal Quote:':  
            quote[l]=k.text.replace("Personal Quote:","")
        
        if k.find('h4',class_='inline').text=='Nickname:':  
            nick[l]=k.text.replace("Nickname:","")
        
        if k.find('h4',class_='inline').text=='Height:':  
            height[l]=(k.text.replace("Height:",""))
        
        if k.find('h4',class_='inline').text=='Alternate Names:':  
            a_names[l]=k.text.replace("Alternate Names:","")
    try:
        j=soup.find('div',class_='image')
    except:
        print("erro")
    if j.find('a'):
        image[l]=(j.find('a').img['src'])
    j=soup.find('div',{"id":"name-born-info"})
    if j:
        dob[l]=(j.find('time').text.rstrip('\n'))
    j=soup.find('div',class_='inline')
    if j:
        bio[l]=(j.text)
    l+=1
    print(l)
```

```python

frame=pd.DataFrame()
frame['name']=name
frame['profile_link']=profile
frame['bio']=bio
frame['D.O.B']=dob
frame['image']=image
frame['Other_Works']=Other_Works
frame['Nick_Name']=nick
frame['star_sign']=star
frame['trivia']=trivia
frame['trademark']=trademark
frame['quote']=quote
frame['height']=height
frame['ALternate Names']=a_names
```
